```python
from spacy.lang.en.stop_words import STOP_WORDS
len(STOP_WORDS)
```

```python
nlp3 = spacy.load("en_core_web_sm")
doc3 = nlp("Sane started celebrating even before Coman touched the ball. He knew Onana wasn't saving that")
for token in doc3:
    if token.is_stop:
        print(token)
```

```python
def preprocess(text):
    doc4 = nlp3(text)
    no_stop_word = [token.text for token in doc4 if not token.is_stop and not token.is_punct]
    #return no_stop_word #if you want list element
    return " ".join(no_stop_word)
```

# TF-IDF :  TF * IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = ["The sun rose over the mountains, painting the sky with streaks of pink and orange.",
          "A gentle breeze rustled the leaves of the trees, sending a shower of golden sunlight dancing to the ground.",
          "A lone bird soared high above, its wings casting a fleeting shadow on the field below.",
#           "A field of wildflowers swayed in the breeze, their vibrant colors a feast for the eyes.",
#           "A babbling brook gurgled over smooth stones, its melody a soothing counterpoint to the chirping of crickets."    
]
```

```python
v = TfidfVectorizer()
transformed = v.fit_transform(corpus)
print(v.vocabulary_)
```

```python
# term which is appear in many doc has lower value
# term which is appear in leat doc or we can say rare term  has higher value

all_feature_name = v.get_feature_names_out()
for word in all_feature_name:
    indx = v.vocabulary_.get(word)
    print(f"{word} : {v.idf_[indx]}")
```

```python
fil = twitt.query("sentiment == 'love' or sentiment == 'fun' or sentiment == 'surprise' or sentiment == 'hate' ")
only = fil[['sentiment', 'content']]

```

```python
only.sentiment.value_counts()
```

```python
def mood(sentiment):
    if sentiment == "love":
        return 0
    elif sentiment == "fun":
        return 1
    elif sentiment == "surprise":
        return 2
    elif sentiment == "hate":
        return 3

only['mood4'] = only['sentiment'].apply(mood)
```

```python
only.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(only.content, only.mood4, test_size = 0.2, random_state = 2022, stratify = only.mood4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('KNN', KNeighborsClassifier())
])
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
y_test[33:39]
```

```python
y_pred[33:39]
```

```python
# With the Multinomial Naive Bayes 


from sklearn.naive_bayes import MultinomialNB

clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('Multi NB', MultinomialNB())
])
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
#With the use of RandomForest Classifier

from sklearn.ensemble import RandomForestClassifier
clf = Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ('Random Forest ', RandomForestClassifier())
])
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

Here we see that the Random Forest classifier algorithm is more accurate than the others.

# Word Embeddings

```python
nlp01 = spacy.load("en_core_web_lg")
```

```python
docs1 = nlp01("dog cat apple monkey egg frm")
for token in docs1:
    print(token.text, "vector : ", token.has_vector, "OOV : ", token.is_oov)
```

```python
docs1[0].vector.shape
```

```python
base_token = nlp01("animal")
```

```python
for token in docs1:
    print(f"{token.text} <-> {base_token.text}: ", token.similarity(base_token))
```

```python
def print_similarity(base_word, words_to_compare):
    base_token = nlp01(base_word)
    docer = nlp01(words_to_compare)
    for token in docer:
        print(f"{token.text} <-> {base_token.text}: ", token.similarity(base_token))
```

```python
print_similarity("monkey", "dog animal april cow man person ")
```
